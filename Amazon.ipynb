## Recommendation Model: Some of the movies hadnâ€™t been watched and therefore, are not rated by the users. Netflix would like to take this as an opportunity and build a machine learning recommendation algorithm which provides the ratings for each of the users.

### 1.Divide the data into training and test data

### 2.Build a recommendation model on training data

### 3.Make predictions on the test data

```python
from sklearn.model_selection import train_test_split
```

```python
amazon=df_copy.melt(id_vars=df_copy.columns[0],value_vars=df_copy[1:],var_name='Movies',value_name='Rating')
amazon
```

```python
!pip install scikit-surprise
```

```python
from surprise import Reader
from surprise import accuracy
from surprise import SVD
from surprise import Dataset
from sklearn.model_selection import train_test_split
```

```python
train_set,test_set=train_test_split(amazon,train_size=0.8)
```

```python
rs=Reader(rating_scale=(-1,10))
train_data=Dataset.load_from_df(train_set.fillna(0),reader=rs)
test_data=Dataset.load_from_df(test_set.fillna(0),reader=rs)
```

```python
test_data=test_data.build_full_trainset()
test_data=test_data.build_testset()
```

```python
model=SVD(n_factors=200,n_epochs=30)
```

```python
train_data=train_data.build_full_trainset()
```

```python
model.train(train_data)
```

```python
pred_test=model.test(test_data)
```

```python
accuracy.rmse(pred_test)
```

```python
u_id='AH3QC2PC1VTGP'
mv = 'Movie206'
r_id = 5.0
model.predict(u_id, mv, r_ui=r_id, verbose= True)
```
