```python
feature_names=['OverallGrade', 'Obedient', 'ResearchScore', 'ProjectScore']
training_feature=df[feature_names]
```

```python
outcome_name=['Recommend']
outcome_label=df['Recommend']
```

```python
numeric_feature_names=['ResearchScore','ProjectScore']
categorical_feature_names=['OverallGrade', 'Obedient']
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit(training_feature[numeric_feature_names])
training_feature[numeric_feature_names]=ss.transform(training_feature[numeric_feature_names])
```

```python
training_feature
```

```python
training_feature=pd.get_dummies(training_feature,columns=categorical_feature_names)
training_feature
```

```python
categorical_engineered_features=list(set(training_feature.columns)-set(numeric_feature_names))
```

```python
categorical_engineered_features
```

```python
from sklearn.linear_model import LogisticRegression
import warnings;warnings.simplefilter('ignore')
```

```python
training_feature
```

```python
lr=LogisticRegression()
model=lr.fit(training_feature,outcome_label)
model
```

```python
prediction_features=new_data[feature_names]
prediction_features[numeric_feature_names]=ss.transform(prediction_features[numeric_feature_names])
prediction_features=pd.get_dummies(prediction_features,columns=categorical_feature_names)
```
